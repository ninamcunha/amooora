# Inspecting Correlations to select features

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ok=pd.read_csv('../raw_data/ok_clean.csv')
topics = pd.read_csv('../raw_data/text_and_topics_classified_encoded_all.csv', keep_default_na=False)
ok = pd.concat([ok, topics], axis=1)
```

```python
list(ok.columns)
```

## Essay columns (text)

### Topics generated from LDA

```python
selected_columns = [
 'topic_0_from_five', 'topic_1_from_five', 'topic_2_from_five', 'topic_3_from_five', 'topic_4_from_five', 
 'topic_0', 'topic_1', 'topic_2', 'topic_3', 'topic_4',
 'topic_0_from_two','topic_1_from_two',
 'topic2_0', 'topic2_1', 'cluster_text'
]

corr = ok[selected_columns]

plt.figure(figsize=(8, 6))
sns.heatmap(corr.corr(), annot=True, cmap='coolwarm', fmt=".1f", linewidths=.5)

plt.title('Correlation LDA generated topics')

# Show the plot
plt.show()
```

### PCA generated topics

```python
selected_columns = [
 'PCA_vect_1','PCA_vect_2',
 'PCA_emb_1','PCA_emb_2',
 'PCA_sdv_1','PCA_sdv_2', 'cluster_text'
]

corr = ok[selected_columns]

plt.figure(figsize=(8, 6))
sns.heatmap(corr.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

plt.title('Correlation PCA generated topics')

# Show the plot
plt.show()
```

```python
def check_feature_statistics(dataset):
    # Get summary statistics for numeric features
    print("Numeric Features Statistics:")
    print(dataset[['age_scaled', 'height_scaled', 'ethnicity_count', 'income_reported', 'non_paying_job']].describe())

    # Get distribution of categorical features
    print("\nCategorical Features Distribution:")
    
    # Loop through each feature set and print distribution for categorical variables
    feature_sets = {
        'basic': ['single', 'female', 'orientation_bisexual', 'orientation_gay', 'orientation_straight'],
        'body_type': ['body_type_athletic', 'body_type_average', 'body_type_curvy', 'body_type_fit', 'body_type_not_disclosed', 
                      'body_type_other', 'body_type_thin'],
        'diet': ['diet_type_anything', 'diet_type_not_disclosed', 'diet_type_other', 'diet_type_vegetarian', 
                 'diet_seriousness_mostly', 'diet_seriousness_not_disclosed', 'diet_seriousness_strictly'],
        'drink': ['drinks_not at all', 'drinks_not_disclosed', 'drinks_often', 'drinks_rarely', 'drinks_socially'],
        'drugs': ['drugs_never', 'drugs_not_disclosed', 'drugs_sometimes'],
        'education': ['education_type_college_univ', 'education_type_grad_or_professional_edu', 
                      'education_type_not_disclosed', 'education_type_two_year_college_or_less', 'education_status_graduated',
                      'education_status_not_disclosed', 'education_status_working'],
        'ethnicity': ['is_asian', 'is_white', 'is_black', 'is_other', 'is_hispanic_latin', 'is_ethnicity_nan'],
        'job': ['job_creative_education', 'job_healthcare_law', 'job_not_disclosed', 'job_stem_business', 'job_student', 
                'job_trades_service', 'job_unemployed_retired_other'],
        'has_kids': ['no_of_kids_more_than_one', 'no_of_kids_not_disclosed', 'no_of_kids_one', 'no_of_kids_zero'],
        'want_kid': ['want_more_kids_maybe', 'want_more_kids_no', 'want_more_kids_not_disclosed', 'want_more_kids_yes'],
        'has_dog': ['has_dogs_no', 'has_dogs_not_disclosed', 'has_dogs_yes']
    }

    # Iterate through feature sets
    for group, features in feature_sets.items():
        print(f"\nDistribution for {group}:")
        for feature in features:
            print(f"\nFeature: {feature}")
            print(dataset[feature].value_counts(normalize=True).sort_index())

# Call the function with your dataset
check_feature_statistics(ok)
```
